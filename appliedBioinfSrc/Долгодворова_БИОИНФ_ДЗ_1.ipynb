#### compseq

```python
for i in range(10):
  seq = f"splitted_file/Record_CP033690_splittedPart{i + 1}.fasta"
  comp = f"compseq_file/Record_CP033690_splittedPart{i + 1}.fasta"
  print(comp)
  !./EMBOSS-6.6.0/emboss/compseq -sequence=$seq -word=4 -outfile=$comp
```

#### соберем результаты

```python
needed = ['AAAA', 'ATAT', 'ATTA', 'AATT']
```

```python
filename, word, obs_count, obs_frequency, exp_frequency, obsexp_frequency = [], [], [], [], [], []
for i in range(10):
  fileName = f"compseq_file/Record_CP033690_splittedPart{i + 1}.fasta"
  with open(fileName) as f:
      fileCompseq = f.readlines()
      for line in fileCompseq:
        if line[0:4] in needed:
          splitLines = line.split()
          word.append(splitLines[0])
          obs_count.append(splitLines[1])
          obs_frequency.append(splitLines[2])
          exp_frequency.append(splitLines[3])
          obsexp_frequency.append(splitLines[4])
          filename.append(fileName)
      f.close()
```

#### Датафрейм содержит нужные нам результаты
можно запринтить весь, но я оставила пока так

```python
import pandas as pd
df_compseq = pd.DataFrame({'word': word,
                   'obs_count': obs_count,
                   'obs_frequency': obs_frequency,
                   'exp_frequency': exp_frequency,
                   'obsexp_frequency': obsexp_frequency,
                   'name': filename
                   })
print(df_compseq.to_markdown())
```

### Сравнение результатов


### среднее, дисперсия и p-value

```python
df_compseqShuffle = df_compseqShuffle.astype({'obs_count':'int'})
```

```python
df_compseq = df_compseq.astype({'obs_count':'int'})
```

#### среднее, дисперсия, p-value

```python
import statistics
import numpy as np
from scipy import stats
```

```python
needed
```

```python
for i in needed:
  print(f"среднее {i} для неперемешанного файла",
        statistics.mean(list(df_compseq.loc[df_compseq['word'] == i]['obs_count'])))

  print(f"среднее {i} для перемешанного (shuffled) файла",
        statistics.mean(list(df_compseqShuffle.loc[df_compseqShuffle['word'] == i]['obs_count'])))

  print(f"дисперсия {i} для неперемешанного файла",
        np.var(list(df_compseq.loc[df_compseq['word'] == i]['obs_count'])))

  print(f"дисперсия {i} для перемешанного (shuffled) файла",
        np.var(list(df_compseqShuffle.loc[df_compseqShuffle['word'] == i]['obs_count'])))

  print(f"p-value: ", stats.ttest_ind(list(df_compseq.loc[df_compseq['word'] == i]['obs_count']),
                list(df_compseqShuffle.loc[df_compseqShuffle['word'] == i]['obs_count'])).pvalue)

  print('------------------------------------------')
```

```python
! sudo apt-get install -y jellyfish
```

```python
import subprocess
k = 4
kmers = {}
```

```python
while True:
  subprocess.run(f"jellyfish count -m  {k} -s 10M generated_fasta/Record_CP033690.1.fasta", shell=True, check=True, text=True)
  uniq = int(subprocess.check_output("jellyfish histo mer_counts.jf | head -n 1 | awk '{print $2}'", shell=True, text=True))
  total = subprocess.check_output("jellyfish histo mer_counts.jf | awk '{print $2}'", shell=True)
  total = str(total)[2:].split('\\n')
  total = total[:-1]
  total = sum([int(i) for i in total])
  kmers[k] = uniq / total
  k += 1
  if kmers[k - 1] >= 0.99:
    k_min = k - 1
    break
```
