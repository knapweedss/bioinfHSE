```python
import re
def gettingWPs(filename):
  result = []
  with open(f'{filename}') as f:
    lines = f.readlines()
  s = 0
  e = 0
  start = 0
  end = 0
  for line in lines:
    if line.startswith('    E-value  score  bias    E-value  score  bias    exp  N'):
      start += (s + 2)
    if line.startswith('  ------ inclusion threshold ------'):
      end += e
    elif (line.startswith('Domain annotation for each sequence (and alignments):') and
          filename == 'rez_file_PF02518.alignment.seed.txt'):
      end += (e - 2)
    s += 1
    e += 1
  for l in lines[start:end]:
    result.extend(re.findall(r'WP_.+? ', l))
  return result
```

```python
PF00512 = gettingWPs('rez_file_PF00512.alignment.seed.txt')
len(PF00512)
```

```python
PF07730 = gettingWPs('rez_file_PF07730.alignment.seed.txt')
len(PF07730)
```

```python
PF02518 = gettingWPs('rez_file_PF02518.alignment.seed.txt')
len(PF02518)
```
