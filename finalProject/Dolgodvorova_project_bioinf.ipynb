# Попробуем сделать для RefSeq

сортируем по повторяющимся рамкам считывания

```python
# находим дублирующиеся ORF
duplicate_orfs = uniprot_right['orf'].value_counts()
duplicate_orfs = duplicate_orfs[duplicate_orfs > 1]
# выбираем ряды с задублированными ORF
duplicated_rows = uniprot_right[uniprot_right['orf'].isin(duplicate_orfs.index)]
# выбираем ряды где ORF совпадает с 1 задубл. значением
result = duplicated_rows[duplicated_rows['orf'] == duplicate_orfs.index[0]]
```

```python
for orf in duplicated_rows['orf'].unique():
    values = duplicated_rows[duplicated_rows['orf'] == orf]
    max_value = values['identity'].max()
    main_index = values[values['identity'] == max_value].index
    delete_indexes = [i for i in values.index if i not in main_index]
    uniprot_right = uniprot_right.drop(delete_indexes)
```

```python
uniprot_right
```

```python
uniprot_right['orf_name'] = uniprot_right['orf'].apply(lambda x: x.split(':')[0])
uniprot_right['coor1'] = uniprot_right['orf'].apply(lambda x: x.split(':')[1])
uniprot_right['coor2'] = uniprot_right['orf'].apply(lambda x: x.split(':')[2])
```

```python
delete_orfs = set(list(uniprot_right.orf_name))
```

```python
sort = []
with open('__all.fa', 'r') as f:
  f = f.read()
  f = f.split('>')
  for el in f:
    if el != '':
      el_ = el.split(':')[0]
      if el_ not in delete_orfs:
        sort.append(el)
```

```python
# сортируем пересекающиеся последовательности
sequences = []
for i in uniprot_right.values:
  r1 = int(i[-2])
  r2 = int(i[-1])
  sequences.append(list(range(r1, r2+1)))
#sequences[0:10]
```

```python
final_sorting = []
for j in sort:
    j_parts = j.split()[0].split(':')
    if len(j_parts) == 3:
        r1, r2 = map(int, (j_parts[1], j_parts[2]))
        for el in sequences:
            seq2 = set(range(min(r1, r2), max(r1, r2) + 1))
            main_seq = set(el) & seq2
            if not main_seq and j not in final_sorting:
                final_sorting.append(j)
    else:
        print(f"беды с вводом")
```

```python
import re
trna_dict = {}
for trna_fragment in File:
    match = re.search(pattern, trna_fragment)
    if match:
        rna_number = match.group(1)
        amino_acid = match.group(5).split()[0]
        anti_codon = match.group(6)
        start_coordinate = int(match.group(3))
        end_coordinate = int(match.group(4))
        trna_dict[f'rna{rna_number}'] = {
            'aminoacid': amino_acid,
            'anticodon': anti_codon,
            'coors': [start_coordinate, end_coordinate]
        }
print('tRNAscan:')
trna_dict
```

посмотрим пересечения

```python
def find_intersection(dict1, dict2):
    intersection_dict = {}

    for key in dict1.keys() & dict2.keys():
        if isinstance(dict1[key], dict) and isinstance(dict2[key], dict):
            nested_intersection = find_intersection(dict1[key], dict2[key])
            if nested_intersection:
                intersection_dict[key] = nested_intersection
        elif dict1[key] == dict2[key]:
            intersection_dict[key] = dict1[key]

    return intersection_dict
```

# Создаем файл GeneBank

```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.SeqFeature import SeqFeature, FeatureLocation
```

```python
with open('fragment10.fasta','r') as F:
  f = F.read()
  f = f.replace('\n','').replace(' ','')
```

```python
# Create a sequence
sequence_string = f
sequence_object = Seq(sequence_string)
```

```python
# Create a record
record = SeqRecord(sequence_object,
                   id='123456789', # random accession number
                   name='fragment10',
                   description='GenBank file')
```

```python
from Bio import Entrez, SeqIO

def extract_info(sample):
    try:
        orf = sample[-3].split('|')[1].split('_')[0]
        coor1, coor2 = sorted(map(int, sample[-2:]))
        strand = 1 if coor1 < coor2 else -1

        Entrez.email = 'your_email@example.com'
        handle = Entrez.efetch(db="protein", id=sample[1], rettype="gb")
        record = SeqIO.read(handle, "genbank")

        identity = sample[3]
        positives = sample[12]
        e_value = sample[10]

        gene = str(record.name)
        if 'Full' in record.description:
            product = str(record.description.split('Full')[1].replace('=','').split(';')[0])
        else:
            product = str(record.description.split('[')[0])

        translation = str(record.seq)
        note = f"{record.annotations['organism']}; identity: {identity}; positives: {positives}; e-value: {e_value}"

        final = {
            'coor1': coor1,
            'coor2': coor2,
            'id': sample[1],
            'gene': gene,
            'product': product,
            'translation': translation,
            'note': note,
            'strand': strand,
            'orf': orf
        }

        return final

    except Exception as e:
        return None
```

```python
final = []
try:
  for i in result_df2.values:
      final.append(extract_info(i))
except:
  pass
```

```python
len(final)
```

```python
final = final[:-6] # удаляем None
```

```python
len(final)
```

```python
# Add annotation
for i in final:
  copy_el = i
  feature = SeqFeature(FeatureLocation(start=copy_el['coor1'],
                                       end=copy_el['coor2'],
                                       strand=copy_el['strand']),
                                       type='CDS',
                      qualifiers={'gene':copy_el['gene'],
                                  'product':copy_el['product'],
                                  'translation':copy_el['translation'],
                                  'note':copy_el['note']})
  record.features.append(feature)
```

```python
# возьмем трнк из tRNAscan

for rna_id, rna_info in trna_dict.items():
    start, end = rna_info['coors']
    if start > end:
        start, end = end, start  # меняем start и end

    product = f"tRNA_{rna_info['aminoacid']}_{rna_info['anticodon']}"

    feature = SeqFeature(FeatureLocation(start=start, end=end),
                         type='tRNA',
                         qualifiers={'product': product})

    record.features.append(feature)
record.annotations["molecule_type"] = "RNA"
```

```python
# Save as GenBank file
output_file = open('result.gb', 'w')
SeqIO.write(record, output_file, 'genbank')
```
